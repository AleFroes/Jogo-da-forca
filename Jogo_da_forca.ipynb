```python

import os
import random


from tqdm import tqdm
import time
for i in tqdm (range(int(10))):
    time.sleep(5)

nome = input ('Digite seu nome')
print (f'Olá seja bem vindo {nome}! Vamos começar a jogar?')
input ('\nPressione enter para começar')

os.system ('clear')

lista_de_palavras = ['gato', 'cachorro', 'vaca', 'leão', 'rato' ,'papagaio', 'tigre', 'elefante', 'cobra']
palavra_selecionada = random.choice (lista_de_palavras).upper()
tamanho_palavra = len(palavra_selecionada)#lê tamanho da palavra
palavra_codificada = ['_']*tamanho_palavra
quantidade_de_erros = 0

while '_' in palavra_codificada and quantidade_de_erros < 6:
  print (f'\nSua palavra tem {tamanho_palavra} letras, o tema é animais')
  print (f'Erros: {quantidade_de_erros} de 6')
  for letra in palavra_codificada:
    print (letra, end = ' ') #end para quebrar a linha
  print()

  letra_escolhida = input ('Digite uma letra: ').upper()
  acertou = False
  for i in range (len(palavra_selecionada)): #lê a palavra e verifica se o usuário acertou alguma letra
     if letra_escolhida == palavra_selecionada [i]: #substitui o underline pela cletra que o usuário acertou
       palavra_codificada [i] = letra_escolhida
       acertou = True

  if acertou == True:
     print ('Parabéns! Acertou.')
  else:
    print ('Errou, essa letra não existe nesta palavra')
    quantidade_de_erros = quantidade_de_erros + 1 #contabiliza erros
if quantidade_de_erros == 6:
  print ('Você perdeu!')
else: print ('Parabéns! Você ganhou!!!')
print (f'A palavra era: {palavra_selecionada}')
```
